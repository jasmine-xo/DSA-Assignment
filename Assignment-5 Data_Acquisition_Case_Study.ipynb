## Q6. Write code to get the list of files in your Downloads folder and save it to a CSV file name "download_list.csv"

```python
import os
import pandas as pd

# Path to Downloads folder
downloads_path = os.path.expanduser("~/Downloads")

# Get list of files (excluding directories)
files = [f for f in os.listdir(downloads_path) if os.path.isfile(os.path.join(downloads_path, f))]

# Create DataFrame
df = pd.DataFrame(files, columns=["Filename"])

# Save to CSV in Downloads folder
output_path = os.path.join(downloads_path, "download_list.csv")
df.to_csv(output_path, index=False)

print("List of files saved to:", output_path)
```

## Q.16 In which season (Spring, Summer, Fall, Winter) people rented bikes the most?

```python

season_map = {1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}
data["season_name"] = data["season"].map(season_map)

# Group by season and calculate total rentals
season_rentals = data.groupby("season_name")["cnt"].sum().sort_values(ascending=False)

print("Total rentals by season:")
print(season_rentals)
print("\nMost rentals occurred in:", season_rentals.idxmax())
```

## Q.17 What is the peak hour in which bike rents the most?

```python
hourly_rentals = data.groupby("hr")["cnt"].sum()

# Find the hour with the highest rentals
peak_hour = hourly_rentals.idxmax()
max_rentals = hourly_rentals.max()

print("Peak Hour:", peak_hour)
print("Total Rentals in Peak Hour:", max_rentals)
```

## Q.18 In which day of a week bikes rents out most?

```python
weekday_map = {
    0: "Sunday", 1: "Monday", 2: "Tuesday", 3: "Wednesday",
    4: "Thursday", 5: "Friday", 6: "Saturday"
}
data["weekday_name"] = data["weekday"].map(weekday_map)

# Group by weekday and sum rentals
weekday_rentals = data.groupby("weekday_name")["cnt"].sum().sort_values(ascending=False)

# Print results
print("Total rentals by weekday:")
print(weekday_rentals)
print("\nDay with most rentals:", weekday_rentals.idxmax())
```

## Q.19 In which hour Casual users rents bikes the most?

```python
# Group by hour and sum casual rentals
hourly_casual = data.groupby("hr")["casual"].sum()

# Find the hour with the highest casual rentals
peak_hour = hourly_casual.idxmax()
peak_rentals = hourly_casual.max()

print("Peak Hour for Casual Users:", peak_hour)
print("Total Casual Rentals at Peak Hour:", peak_rentals)
```

## Q.20 What is the maximum temperature observed in each of the seasons?

```python
# Map season codes to names
season_map = {1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}
data["season_name"] = data["season"].map(season_map)

# Find max temperature per season
max_temp_by_season = data.groupby("season_name")["temp"].max()

print("Maximum temperature observed in each season:")
print(max_temp_by_season)
```
