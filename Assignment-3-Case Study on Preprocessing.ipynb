### Outlier Removal

```python
#boxplot before removing the outliers
df_std_scaled.boxplot()
plt.xticks(rotation=90)
plt.show()
```

```python
# Perform an outlier detection analysis on numerical variables (e.g., using the IQR method).
#except no of floors and age of house, remianing all have outliers.
df_std_scaled
def outlier_removal(df1, col_name):
    q1 = df1[col_name].quantile(0.25)
    q3 = df1[col_name].quantile(0.75)
    iqr = q3 - q1
    upper_bound = q3 + 1.5 * iqr
    lower_bound = q1 - 1.5 * iqr
    df1[col_name] = df1[col_name].clip(lower=lower_bound, upper=upper_bound)
    return df1[col_name]
```

```python
for col in df_std_scaled:
    df_std_scaled[col] = outlier_removal(df_std_scaled, col)
print(df_std_scaled[col])
```

```python
# boxplot after removing the outliers
df_std_scaled.boxplot()
plt.title("Boxplot after Outlier Clipping")
plt.xticks(rotation=90)
plt.show()
```
