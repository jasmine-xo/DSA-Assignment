```python
#3. Statistical Analysis 
#Compute the 25th, 50th, and 75th percentiles of the weight column using NumPy. 
```

```python
#4. Array Manipulation 
#Convert the acceleration column into a NumPy array and normalize its values (scale between 0 and 1).
```

```python
#Convert into array
acc_arr=df["acceleration"].values
#Min value
acc_min=acc_arr.min()
#Max value
acc_max=acc_arr.max()
print("Min value =",acc_min,"Max value =",acc_max)
#Range=max-min value
acc_range=acc_max-acc_min
print("Range=",acc_range)
#Normalization=(value-min)/range
acc_norm=((acc_arr-acc_min)/acc_range)
print("Normalized value =",acc_norm)
```

```python
#5. Broadcasting 
#Increase all horsepower values by 10% and store the updated values in a new NumPy array.
#Handle missing data (if any) by replacing it with the mean of the column before applying the increase. 
```

```python
#Convert to numeric (turns '?' into NaN)
horse_arr = pd.to_numeric(df['horsepower'], errors='coerce').values
# Calculate mean horsepower (ignoring NaN)
horse_mean = np.nanmean(horse_arr)
# Replace NaN values with the mean
horse_replace = np.where(np.isnan(horse_arr), horse_mean, horse_arr)
horse_increase = horse_replace * 1.10
print(horse_increase)
```

```python
#7. Matrix Operations 
#Create a 2D NumPy array containing the columns mpg, horsepower, and weight. 
#Compute the dot product of this matrix with a given vector [1, 0.5, -0.2]. 
```

```python
numpy_arr=pd.to_numeric(df['horsepower'], errors='coerce').values
weight=df["weight"].values
mpg=df["mpg"].values
stack=np.column_stack((mpg,numpy_arr,weight))
print("Numpy array",stack)
vector=np.array([1, 0.5, -0.2])
dot_pdt=np.dot(stack,vector)
print("Dot product =",dot_pdt)
```

```python
#10. Conditional Aggregates 
#Calculate the mean mpg for cars grouped by the number of cylinders using NumPy techniques.
```

```python
cylinders = df['cylinders'].to_numpy()
mpg = df['mpg'].to_numpy()
group_cylinders = np.unique(cylinders)
for i in group_cylinders:
    mean_mpg=mpg[cylinders == i].mean()
    print("MPG Mean of cylinder",i,"=",mean_mpg)
```

```python
#3. Handling Missing Data 
#Identify if there are any missing values in the dataset. Replace missing values in the horsepower column with the column's median. 
```

```python
df.isna().sum()
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
print("After replaciing missing values",df.isna().sum())
```

```python
#4. Data Transformation 
#Add a new column power_to_weight_ratio, calculated as horsepower / weight.
```

```python
df['horsepower'] = df['horsepower'].replace('?', np.nan)
df['horsepower'] = df['horsepower'].astype(float)
df["power_to_weight_ratio"]=df["horsepower"]/df["weight"]
df
```

```python
#10. Finding Anomalies 
#Identify potential outliers in the mpg column using the Interquartile Range (IQR) method. Specifically: 
#● Calculate the IQR for mpg. 
#● Define outliers as values less than Q1 - 1.5 * IQR or greater than Q3 + 1.5 * IQR. 
#● Create a DataFrame of cars classified as outliers, displaying car_name, mpg, and model_year. 
```

```python
describe=df.describe()
Q1=describe.loc["25%","mpg"]
Q3=describe.loc["75%","mpg"]
IQR=Q3-Q1
print("IQR =",IQR)
Lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print("Lower bound =",Lower_bound,",Upper bound =",upper_bound)
outliers=df[(df["mpg"] < Lower_bound) | (df["mpg"] >   upper_bound)][["car name","mpg","model year"]]
outliers
```
