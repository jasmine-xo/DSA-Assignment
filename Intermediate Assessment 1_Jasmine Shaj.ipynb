### Part 2: Data Cleaning (30-40 min)

6. Check for and remove any duplicated rows.

```python
df_new = df.drop_duplicates()  #To delete duplicates
df_new
```

```python
duplicates = df.duplicated().sum() # to check duplicates
duplicates
```

7. Handle missing values if any:
○ Fill missing numerical features with the mean value

```python
df = df.fillna(df.mean())
```

```python
df
```

8. Create a new column PricePerRoom = median_house_value / total_rooms.

```python
df["PricePerRoom"] = df["MedHouseVal"]/df["AveRooms"]
df
```

9. Create a column HighPopulationArea:
○ 1 if population > 500, else 0.

```python
df["HighPopulationArea"]= df["Population"].apply(lambda x:1 if x>500 else 0)
df["HighPopulationArea"]
```

```python
df
```

10. Bin the median_income into 5 equal-sized bins and label them as Very Low, Low, Medium,
High, Very High. (Hint: pd.cut)

```python
labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']
df['income_category'] = pd.cut(df['MedInc'], bins=5, labels=labels)
```

```python
df
```

11. Drop columns that seem redundant after feature creation (if any).

```python
df = df.drop(columns=['MedInc'])
df
```

### Part 3: Data Visualization (40-50 min)

12. Plot the distribution of median_house_value with a histogram.

```python
import matplotlib.pyplot as plt
plt.hist(df["MedHouseVal"], color="purple", rwidth=0.9, density=True)
plt.title("Histogram of median_house_value")
plt.xlabel("MedHouseVal")
plt.ylabel("Density")
plt.show()
```

13. Create a scatter plot of longitude vs latitude, colored by median_house_value.

```python
plt.figure(figsize=(9, 7))
plt.scatter(df["Longitude"], df["Latitude"], c=df["MedHouseVal"])
plt.title("Longitude vs Latitude")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

14. Plot a boxplot of median_house_value grouped by the new income categories.

```python
df.boxplot(column='MedHouseVal', by='income_category', figsize=(11,6))
plt.title('Median House Value by Income Category')
plt.xlabel('income_Category')
plt.ylabel('MedHouseVal')
plt.show()
```

15. Plot the correlation matrix heatmap between numerical features.

```python
import seaborn as sns
plt.figure(figsize=(10, 7))
data_new=df[["HouseAge","AveRooms","AveBedrms","Population","AveOccup","MedHouseVal","PricePerRoom","HighPopulationArea","Longitude","Latitude"]]
sns.heatmap(data_new.corr(), annot=True)
plt.xticks(rotation=45)
```

16. Create a bar plot showing average median_house_value for high population vs low
population areas.

```python
avg_price = df.groupby('HighPopulationArea')['MedHouseVal'].mean()
avg_price.plot(kind='bar', figsize=(6,4), color=['yellow', 'green'])
plt.title('Average Median House Value\nHigh vs Low Population Areas')
plt.xlabel('High Population Area (0 = Low, 1 = High)')
plt.ylabel('Average Median House Value')
plt.show()
```
